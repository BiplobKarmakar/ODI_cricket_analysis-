## Meaning of each column

Column Name | Meaning
--- | ---
**Player** | **Name of the Player**
**Span**|**Number of years a player played for**
**Mat**|**Total Number of matches a player played for**
**Inns**|**Number of matches a player bowled for**
**Balls**|**Number of years a player played for**
**Runs**|**Number of Runs a player had given**
**Wkts**|**Number of Wickets a player hunted**
**Ave**|**Average runs spent for a wicket**
**Econ**|**Runs spent per Over**
**SR**|**Bowling Strike Rate is a measurement of a bowler’s average number of balls bowled for every wicket taken.**
**4**|**Total number of times the bowler takes 4 wickets**
**5**|**Total number of times the bowler takes 5 wickets**
