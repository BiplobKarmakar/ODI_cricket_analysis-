**7. Rename the column names appropriately**


```python
df = df.rename(columns={'Mat':'Matches', 
                        'Inns':'Innings',
                        'Wkts': 'Wickets',
                        'Econ': 'Economy Rate',
                        'Ave': 'Average',
                       # 'BF': 'Balls_Faced',
                        'SR': 'Strike_Rate',
                        4:'Four_wickets',
                        5:'Five_wickets'
                        
                        })



new_col_sequence = ['Player', 'Span', 'Matches', 'Innings', 'Balls', 'Runs', 'Wickets',
      'Average','Economy Rate', 'Strike_Rate', 'Four_wickets', 'Five_wickets']
df = df[new_col_sequence]

display(df.head())


```

**8. How many players played for ICC?**


```python
df[["Player_Name", "Country"]] = df['Player'].str.split("(", expand=True)

df = df.drop('Player', axis=1)

df['Country'] = df['Country'].str.replace(")", "")

new_col_sequence = ['Player_Name','Country', 'Span', 'Matches', 'Innings', 'Balls', 'Runs', 'Wickets',
      'Average','Economy Rate', 'Strike_Rate', 'Four_wickets', 'Five_wickets']
df = df[new_col_sequence]

display(df.head())
```

```python
def icc_check(x):
    if "ICC" in x:
        return "Yes"
    else:
        return "No"
```

```python
df['played_for_ICC'] = df['Country'].apply(icc_check) 
display(df.head(10))

print(df['played_for_ICC'].value_counts())
```

**9. How many different countries are present in this dataset?** 


```python
df['Country'] = df['Country'].str.replace("ICC/", "")
df['Country'] = df['Country'].str.replace("/ICC", "")
df['Country'] = df['Country'].str.replace("Asia/", "")
df['Country'] = df['Country'].str.replace("Afr/", "")
df['Country'] = df['Country'].str.replace("ENG/", "")



df['Country'].value_counts()
```

**10. Which player(s) had played for the longest period of time?**


```python
df[['start_year', 'end_year']] = df['Span'].str.split("-", expand=True)

# removing the "Span" column
df = df.drop("Span", axis=1)

#display(df.head())
df['start_year'] = df['start_year'].astype('int') 
df['end_year'] = df['end_year'].astype('int')

#print(df.info())
df['years_played'] = df['end_year'] - df['start_year']+1

df = df.drop(['start_year', "end_year"], axis=1)

display(df.head(10))
```

```python
df.sort_values(by='years_played', ascending = False).head(10)
```
