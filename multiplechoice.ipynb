```python
friends_s1 = pd.DataFrame()
friends_s1["character"] = ["Rachel Green", "Ross Geller", "Phoebe Buffey","Joey Tribbiani","Monica geller","Chantler Bing"]
friends_s1["Name"] = ["Jenifer Anistin", "David Schwimmer", "Lisa Kudr","Matt Leblanc","Courteney Cox","Mathew Perry"]
friends_s1["Birthday"] = ["1969-02-11", "1966-11-02", "1963-07-30", "1967-07-25","1964-06-15","1969-08-19"]

display(friends_s1)
# print(type(df_A))
print(friends_s1.shape)
```

```python
friends_s2 = pd.DataFrame()
friends_s2["character_name"] = ["Rachel Green", "Ross Geller", "Phoebe Buffey","Joey Tribbiani","Monica geller","Chantler Bing","Richard Buske","Gunther","Emily Waltham","Mike Hannigan"]
#friends_s2["Name"] = ["Jenifer Anistin", "David Schwimmer", "Lisa Kudr","Matt Leblanc","Courteney Cox","Mathew Perry"]
friends_s2["Gender"] = ["Female","Male","Female","Male","Female","Male","Male","Male","Female","Male"]

display(friends_s2)
# print(type(df_A))
print(friends_s2.shape)
```
